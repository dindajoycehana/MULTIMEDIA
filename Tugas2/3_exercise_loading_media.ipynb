# Pendahuluan

## Apa itu Data Multimedia?

Data multimedia adalah informasi yang dikodekan dalam berbagai format untuk merepresentasikan dunia nyata:

- **Audio (1D)**: Sinyal satu dimensi yang berubah terhadap waktu
  - Contoh: musik, suara, speech
  - Representasi: amplitudo vs waktu
  
- **Gambar (2D)**: Matriks nilai intensitas dalam ruang dua dimensi
  - Contoh: foto, ilustrasi, grafik
  - Representasi: intensitas pixel pada koordinat (x,y)
  
- **Video (2D + Waktu)**: Rangkaian frame (gambar) yang ditampilkan berurutan
  - Contoh: film, rekaman, animasi
  - Representasi: frame berubah terhadap waktu dengan frame rate tertentu

### A2. TODO: Muat & Metadata

**Instruksi:**
Tulis kode untuk memuat file audio dan menampilkan metadata dasar:
- Sample rate (Hz)
- Durasi (detik)
- Jumlah kanal (mono/stereo)
- Jumlah total sampel

**Catatan:** Jika file MP3 bermasalah saat loading, gunakan format WAV sebagai alternatif.


```python
### A2. Todo : Muat &  Metadata
# Muat file audio dan tampilkan metadata dasar
audio_path = 'data/jazz_instrument.mp3'  # Ganti dengan path file audio Anda

# Membaca file audio (sf.read bisa membaca mp3 jika backend mendukung, jika error gunakan wav)
data, samplerate = sf.read(audio_path)
duration = len(data) / samplerate

# Jumlah kanal: jika data 1D = mono, jika 2D = stereo/multi
if data.ndim == 1:
    channels = 1
    channel_type = "Mono"
else:
    channels = data.shape[1]
    channel_type = "Stereo" if channels == 2 else f"{channels} channels"

print("Metadata Audio")
print("=" * 40)
print(f"Sample rate: {samplerate} Hz")
print(f"Durasi: {duration:.2f} detik")
print(f"Jumlah kanal: {channels} ({channel_type})")
print(f"Jumlah total sampel: {data.shape[0]:,}") # Total sampel = durasi * sample rate
print()
print("Statistik Amplitudo")
print("=" * 40)
print(f"Nilai minimum: {np.min(data):.4f}")
print(f"Nilai maksimum: {np.max(data):.4f}")
print(f"Nilai rata-rata: {np.mean(data):.4f}")
print(f"Standar deviasi: {np.std(data):.4f}")
```

### A3. TODO: Waveform

**Instruksi:**
Plot waveform audio dengan:
- Sumbu X: waktu (detik)
- Sumbu Y: amplitudo
- Judul dan label sumbu yang jelas

```python
#plot waveform audio
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

# Plot 1: waveform audio Lengkap
times = np.arange(len(data)) / samplerate
ax1.plot(times, data, color='blue')
ax1.set_title("Waveform Audio Lengkap: Jazz Instrumental")
ax1.set_xlabel("Waktu (detik)")
ax1.set_ylabel("Amplitudo")
ax1.set_xlim([0, times[-1]])

# Plot 2: waveform audio 30-second pertama
max_duration = 30  # detik
max_samples = int(samplerate * max_duration)
data_30sec = data[:max_samples]
times_30sec = np.arange(len(data_30sec)) / samplerate

ax2.plot(times_30sec, data_30sec, color='blue')
ax2.set_title("Waveform Audio (30 Sec Pertama): Jazz Instrumental")
ax2.set_xlabel("Waktu (detik)")
ax2.set_ylabel("Amplitudo")
ax2.set_xlim([0, min(max_duration, times_30sec[-1])])

plt.tight_layout()
plt.show()
```

**Analisis yang diperlukan:**
Jelaskan apa yang Anda lihat dari waveform (pola amplitudo, bagian keras/pelan, dll.)

**Jawaban Analisis:**
Berdasarkan gambar waveform diatas, menunjukkan bentuk amplitudo gelombang hampir rata dan padat, berkisar -+(0.5-1). Akan tetapi pada detik direntang 120-130 dan mendekati akhir audio, gelombang tampak mengecil. Mungkin part yang gelombangnya kecil ini berisi jeda lagu, transisi, ending lagu, yang berupa alunan musik bervolume rendah. ### A4. TODO: Spectrogram log-dB

**Instruksi:**
Hitung STFT dan tampilkan spectrogram dalam skala log-dB:
- Gunakan parameter standar (n_fft=1024, hop_length=256)
- Tampilkan dengan colorbar
- Label sumbu: waktu (detik) dan frekuensi (Hz)

```python
### A4. Todo: Spectrogram log-dB
n_fft = 1024  # Panjang jendela FFT 
hop_length = 256  # Jarak antar jendela 

# Karena audio stereo, konversi ke mono dengan merata-rata kedua channel
y = data if data.ndim == 1 else np.mean(data, axis=1)

spectrogram = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
spectrogram_db = librosa.amplitude_to_db(np.abs(spectrogram), ref=np.max)

plt.figure(figsize=(14, 5))
librosa.display.specshow(spectrogram_db, y_axis='log', x_axis='time', sr=samplerate, hop_length=hop_length)
plt.colorbar(format='%+2.0f dB')
plt.title("Spectrogram: Jazz Instrumental (Seluruh Audio)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Frekuensi (Hz)")
plt.tight_layout()
plt.show()

```

**Analisis yang diperlukan:**
Jelaskan perbedaan informasi yang didapat dari spectrogram dibanding waveform.

**Jawaban Analisis:**
Pada waveform kita dapat melihat tinggi rendahnya suara audio, hal ini berkaitan dengan volume keras/pelan dari audio tersebut. Berbeda halnya dengan spectogram, yang menunjukkan frekuensi dari audio tersebut, dimana semakin gelap warnanya, maka itu merupakan suara frekuensi rendah,dan sebaliknya jika berwarna terang, artinya frekuensi tinggi. 

Berdasarkan gambar spectogram diatas, audio didominasi oleh suara berfrekuensi rendah direntang 50-2000Hz , dapat dilihat kepadatan warna kuning terang pada gambar. Spectogram ini menunjukkan bahwa audio kemungkinan diisi oleh suara alat musik seperti bass atau drum. Namun suara berfrekuensi tinggi juga cukup padat memenuhi audio. Kemungkinan ini merupakan suara dari instrumen lainnya pada musik jazz, seperti piano, Saxo,gitar, ataupun gemerincing (tipis).

Pada suara berferekunsi tinggi, masih sering bercampur dengan suara berfrekuensi rendah. Kemungkinan karena musik jazz memainkan beragam musik di waktu bersamaan, sehingga warnanya bercampur terang dan gelap.

### A5. TODO: MFCC

**Instruksi:**
Hitung dan tampilkan minimal 13 koefisien MFCC sebagai heatmap:
- Sumbu X: waktu (frame)
- Sumbu Y: koefisien MFCC (1-13)
- Gunakan colorbar dan judul yang jelas

```python
### A5. Todo: MFCC
# Hitung dan tampilkan minimal 13 koefisien MFCC sebagai heatmap untuk seluruh audio

# Karena audio stereo, konversi ke mono dengan merata-rata kedua channel
y = data if data.ndim == 1 else np.mean(data, axis=1)

mfccs = librosa.feature.mfcc(y=y.astype(float), sr=samplerate, n_mfcc=13)

plt.figure(figsize=(12, 6))
im = plt.imshow(mfccs, aspect='auto', origin='lower', cmap='viridis')
plt.title("MFCC (13 Koefisien) - Jazz Instrumental (Seluruh Audio)")
plt.xlabel("Waktu (frame)")
plt.ylabel("Koefisien MFCC (1-13)")
plt.colorbar(im, label="Nilai MFCC")
plt.tight_layout()
plt.show()
```

**Analisis yang diperlukan:**
Interpretasi sederhana: apakah pola MFCC stabil atau berubah-ubah? Apa potensi maknanya?

**Jawaban Analisis:**
Berdasarkan visualisasi MFCC diatas, menunjukkan bahwa pola MFCC pada koefisien 1-3 cenderung stabil dengan warna kuning terang. Akan tetapi frekuensi >4 cenderung mengalami perubahan warna hijau-biru-kuning(gradasi). Mungkin itu berupa perubahan sedikit terhadap dinamika atau variasi permainan.

### B4. TODO: Histogram Warna

**Instruksi:**
Tampilkan histogram distribusi intensitas untuk channel R, G, B:
- Range: 0-255
- Plot terpisah atau overlay dengan warna sesuai channel
- Label sumbu: intensitas pixel dan frekuensi
- Legend yang jelas

```python
### B4. Todo: Histogram Warna
# Histogram distribusi intensitas untuk channel R, G, B

image_path = 'data/foto_KKN.JPG' 

# Membaca gambar dan konversi ke RGB
image_bgr = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
channels = ['Red', 'Green', 'Blue']
colors = ['r', 'g', 'b']

for i, color in enumerate(colors):
    hist = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=color, label=channels[i])
    plt.xlim([0, 255])

plt.title("Histogram Intensitas Warna (RGB) - Foto di Posko KKN")
plt.xlabel("Intensitas Pixel (0-255)")
plt.ylabel("Frekuensi")
plt.legend()
plt.tight_layout()
plt.show()
```

- Di rentang pixel 0-50, setiap pixel memiliki RGB yang hampir sama rata frekuensinya, kemungkinan ini menandakan gambar berwarna gelap. 

- Selanjutnya dari 50-150, channel biru lebih mendominasi, kemungkinan dalam rentang ini berisi warna-warna yang tidak terlalu terang maupun terlalu gelap(sedang), mungkin seperti banner atau warna baju. Walau ada sedikit perubahan di rentang 130-150 yaitu hijau mendominasi, bisa saja karena warna rumput. 

- Kemudian di rentang 150-250 di dominasi warna merah dan hijau, kemungkinan ini adalah warna-warna coklat/sedikit pucat. Kemungkinan ini adalah warna-warna yang cukup terang, seperti warna dinding rumah ataupun vest kkn. Dan terakhir ada lonjakan tinggi dari 250 keatas terhadap warna biru, merah, dan hijau, ini menandakan ada warna yang sangat terang, kemungkinan adalah langit yang sudah hampir putih.

### C2. TODO: Baca & Metadata

**Instruksi:**
Baca video dengan OpenCV dan tampilkan metadata:
- Resolusi (Width √ó Height)
- Frame rate (fps)
- Jumlah total frame
- Durasi (detik)
- Klasifikasi resolusi (HD, Full HD, 4K, dll.)


```python
### C2. Todo: Baca & Metadata Video
# Baca video dan tampilkan metadata

video_path = 'data/Video_perkenalan_dinda.mp4' 

cap = cv2.VideoCapture(video_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
duration = frame_count / fps if fps > 0 else 0

# Klasifikasi resolusi
if width >= 3840 and height >= 2160:
    res_class = "4K"
elif width >= 1920 and height >= 1080:
    res_class = "Full HD"
elif width >= 1280 and height >= 720:
    res_class = "HD"
else:
    res_class = "SD/Low"

print("Metadata Video")
print("=" * 40)
print(f"Resolusi: {width} x {height} (Width x Height)")
print(f"Frame rate: {fps:.2f} fps")
print(f"Jumlah total frame: {frame_count}")
print(f"Durasi: {duration:.2f} detik")
print(f"Klasifikasi resolusi: {res_class}")

cap.release()
```

**Analisis yang diperlukan:**

**Jawaban Analisis:**

### C3. TODO: Tampilkan 3 Frame (Awal‚ÄìTengah‚ÄìAkhir)

**Instruksi:**
Ambil dan tampilkan 3 frame representatif:
- Frame pertama (index 0)
- Frame tengah (index ~total_frame/2)
- Frame terakhir (index total_frame-1)
- **Konversi BGR‚ÜíRGB** sebelum ditampilkan
- Subplot dengan judul frame dan timestamp


```python
### C3. Todo: Tampilkan 3 Frame (Awal‚ÄìTengah‚ÄìAkhir)
# Ambil dan tampilkan 3 frame representatif dari video (awal, tengah, akhir)
video_path = 'data/Video_perkenalan_dinda.mp4'

cap = cv2.VideoCapture(video_path)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Indeks frame yang diambil
indices = [0, frame_count // 2, frame_count - 1]
frames = []
timestamps = []

for idx in indices:
    cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
    ret, frame = cap.read()
    if ret:
        # Konversi BGR ke RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames.append(frame_rgb)
        # Hitung timestamp (detik)
        timestamp = idx / fps if fps > 0 else 0
        timestamps.append(timestamp)
    else:
        frames.append(None)
        timestamps.append(None)

cap.release()

# Tampilkan frame dengan subplot
plt.figure(figsize=(16, 6))
titles = [
    f"Frame Awal (Index 0)\nTimestamp: {timestamps[0]:.2f} detik",
    f"Frame Tengah (Index {indices[1]})\nTimestamp: {timestamps[1]:.2f} detik",
    f"Frame Akhir (Index {indices[2]})\nTimestamp: {timestamps[2]:.2f} detik"
]

for i, (frame, title) in enumerate(zip(frames, titles)):
    plt.subplot(1, 3, i + 1)
    if frame is not None:
        plt.imshow(frame)
        plt.title(title)
        plt.axis('off')
    else:
        plt.text(0.5, 0.5, "Frame tidak tersedia", ha='center', va='center')
        plt.axis('off')

plt.suptitle("3 Frame Representatif dari Video (Awal, Tengah, Akhir)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

**Analisis yang diperlukan:**
Deskripsikan perbedaan visual antar frame dan apa yang dapat dipelajari dari sampel frame ini.

**Jawaban Analisis:**
- Pada frame 1 ada bagian gambar yang blur/bersambung, yaitu bentuk lambaian tangan (gerakan cepat). 
- Pada frame 2 juga masih terlihat sedikit bagian blur pada jari jempol bawah. Dan bagian atas gambar tampak lebih cerah akibat cahaya lampu. Berefek juga pada tingkat kecerahan gambar yaitu jadi sedikit lebih cerah
- pada frame 3 frame lebih stabil, tidak ada terlihat bagian blur.

Kemungkinan penyebab blur pada gambar dipengaruhi oleh fps 30, sehingga gerakannya lebih patah-patah karena frame nya hanya mengcapture 30 frame per second. Semakin banyak fps, hasil video akan semakin mulus, karena frame yang dicapture lebih banyak, sehingga pergerakannya lebih detail.
